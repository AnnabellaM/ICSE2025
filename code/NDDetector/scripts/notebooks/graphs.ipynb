```python
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib.legend_handler import HandlerTuple


df = pd.read_csv("ICSE2025_AGGREGATE_DATA.csv")
df['%consistent_result'] = df['%consistent_result'] * 100
df = df.sort_values('id')
df
```

```python
plt.figure(figsize=(10, 6))

custom_colors = ["#66c2a5", "#ccece6", "#fc8d62", "#8da0cb", "#c6dbef", "#e78ac3", "#f3c7d5", "#a6d854", "#ffd92f", "#e5c494", "#b3b3b3"]
ax = sb.boxplot(data=df, patch_artist=True, x='configuration', y='%consistent_result', hue='tool', dodge=False, palette=custom_colors)
sb.stripplot(data=df, x='configuration', y='%consistent_result', ax=ax, dodge=False, hue='tool', palette=custom_colors, jitter=False)
handles, labels = ax.get_legend_handles_labels()

ax.legend(handles=[
    (handles[0], handles[11]),
    (handles[1], handles[12]),
    (handles[2], handles[13]),
    (handles[3], handles[14]),
    (handles[4], handles[15]),
    (handles[5], handles[16]),
    (handles[6], handles[17]),
    (handles[7], handles[18]),
    (handles[8], handles[19]),
    (handles[9], handles[20]),
    (handles[10], handles[21])
    ],
            loc='lower center', handlelength=4,
    labels=['Amandroid-I', 'Amandroid-II', 'Infer', 'FlowDroid-I', 'FlowDroid-II', 'Soot-I', 'Soot-II', 'DOOP', 'OPAL', 'Code2Flow', 'PyCG'],
          handler_map={tuple: HandlerTuple(ndivide=None)}, ncol=5, bbox_to_anchor=(0.5, 1.0))
ax.set(xticklabels=[])
ax.set_xlabel('Configurations', fontsize=16)
ax.set_ylabel('% Consistent Results', fontsize=16)
ax.grid(False)
ax
```

```python
plt.figure(figsize=(10, 6))
custom_colors = ["#66c2a5", "#ccece6", "#fc8d62", "#8da0cb", "#c6dbef", "#e78ac3", "#f3c7d5", "#a6d854", "#ffd92f", "#e5c494", "#b3b3b3"]
ax = sb.boxplot(data=df, x='configuration', y='distribution', hue='tool', dodge=False, palette=custom_colors)
sb.stripplot(data=df, x='configuration', y='distribution', hue='tool', ax=ax, dodge=False, palette=custom_colors)
handles, labels = ax.get_legend_handles_labels()
tool_order = ['Amandroid-I', 'Amandroid-II', 'Infer', 'FlowDroid-I', 'FlowDroid-II', 'Soot-I', 'Soot-II', 'DOOP', 'OPAL', 'Code2Flow', 'PyCG']

new_labels = [tool for tool in tool_order]

ax.legend(handles=[
    (handles[0], handles[11]),
    (handles[1], handles[12]),
    (handles[2], handles[13]),
    (handles[3], handles[14]),
    (handles[4], handles[15]),
    (handles[5], handles[16]),
    (handles[6], handles[17]),
    (handles[7], handles[18]),
    (handles[8], handles[19]),
    (handles[9], handles[20]),
    (handles[10], handles[21])
    ],
          loc='lower center', handlelength=4, labels=new_labels,
          handler_map={tuple: HandlerTuple(ndivide=None)}, ncol=5, bbox_to_anchor=(0.5, 1.0))
ax.set(xticklabels=[], yticks=range(2, 7))
ax.set_xlabel('Configurations', fontsize=16)
ax.set_ylabel('# of Distinct Results', fontsize=16)
ax.grid(False)
ax
```
